```python
tasks = {
    "wash dishes": "pending",
    "vacuum floor": "pending",
    "do laundry": "pending",
    "take out trash": "pending"
}

def show_tasks():
    print("\n Current Tasks:")
    for task, status in tasks.items():
        print(f"{task.title()} : {status}")
    print()

def task_agent():
    print("Household Task Agent: Ready to assist! Type a task name to mark it as done. Type 'exit' to stop.")

    while True:
        show_tasks()
        user_input = input("You: ").lower()

        if user_input == "exit":
            print("Agent: Shutting down.")
            break

        if user_input in tasks:
            if tasks[user_input] == "pending":
                tasks[user_input] = "done"
                print(f"Agent: Task '{user_input}' marked as done.\n")
            else:
                print(f"Agent: Task '{user_input}' is already done.\n")
        else:
            print("Agent: Task not recognized. Please type one from the list.\n")

        # Check if all tasks are done
        if all(status == "done" for status in tasks.values()):
            show_tasks()
            print("Agent: All tasks completed!")
            break

task_agent()
```
