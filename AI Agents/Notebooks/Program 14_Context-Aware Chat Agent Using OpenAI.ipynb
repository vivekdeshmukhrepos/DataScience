```python
import openai
import os
```

```python
client= openai.OpenAI(api_key=secret_Key)
chat_history = []
```

```python
def context_agent(user_input):
  chat_history.append({"role": "user", "content": user_input})

  def fallback_response():
   if "how are you" in user_input.lower():
    return "I'm doing well, Thanks!"
   elif "what we did talked about" in user_input.lower():
    previous = [msg["content"] for msg in chat_history[:-1] if msg ["role"] == "user" ]

    if previous:
      return "So far we have discussed: " + ", ".join(chat_history[:-1])
    else:
      return "We have just started chatting"
   else:
    return "Interesting, tell me more"

  try:
    messages = [
        {"role": "system", "content": "You are a helpful assistant."}
    ]+ chat_history

    response = client.chat.completions.create(
      model="gpt-4o-mini",
      messages= [
              {"role": "system", "content": "You are an assistant who uses tools if needed."},
              {"role": "user", "content": user_input}
      ]
    )

    assistamt_reply =  response.choices[0].message.content.strip()
    chat_history.append({"role": "assistant", "content": assistamt_reply})

    return assistamt_reply
  except Exception as e:
    print(e)
    return fallback_response

print("Context-Aware Agent Using GPT-4o Mini")
print("Ask questions like 'How are you?' or 'What did we talk about?'. Type 'exit' to quit.")

while True:
    user_input = input("You: ")
    if user_input.strip().lower() == "exit":
        print("Agent: Goodbye. The session is now ending.")
        break
    response = context_agent(user_input)
    print("Agent:", response)


```
