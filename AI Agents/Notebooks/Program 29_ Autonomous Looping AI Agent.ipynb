```python
import json

secret_file_path = "/content/drive/My Drive/Secret_Keys/OpenAI_Secret_Key.json"

with open(secret_file_path, "r") as f:
  data=json.load(f)

secret_Key = data["open_ai_key"]
```

```python
import openai
import difflib
import time

openai.api_key = secret_Key
```

```python
print("=== Autonomous looping agent ===")
original_sentence = input("Enter a sentence to improve: ").strip()

similarity_threashold = 0.85
max_atempts = 5
temperature = 0.7

for attempt in range(1, max_atempts + 1):
  print("\n == Attempt ", attempt, "---")

  prompt = f"Rephrase the following sentence to make it more unique, clearer or stylistically improved, while keeping the origional meaning:\n\n{original_sentence}"

  response = openai.chat.completions.create(
      model="gpt-4o-mini",
        messages= [
                {"role": "system", "content": "You are a language improvement agent."},
                {"role": "user", "content": prompt}
        ],
      temperature = temperature
  )

  new_sentence = response.choices[0].message.content.strip()

  similarity = difflib.SequenceMatcher(None, original_sentence.lower(), new_sentence.lower()).ratio()

  print("Origional sentence :", original_sentence)
  print("Rephrased sentence:", new_sentence)
  print("Similarity (0 to 1) :", round(similarity, 4))

  if similarity < similarity_threashold:
    print("Result Acceted: The sentence is sufficiently rephrased")
    break
  else:
    print("Result rejected: The sentence is too similar. Retrying ")
    time.sleep(1)

  print("\n ==== Final output ====")
  print("Improved sentence:", new_sentence)
```
