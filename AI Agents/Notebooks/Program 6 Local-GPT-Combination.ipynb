```python
import openai
import os
```

```python
client= openai.OpenAI(api_key=secret_Key)
```

```python
def calculator_tool(expr):
  try:
    return f"Calculated: {eval(expr)}"
  except:
    return "Invalid math expression."
```

```python
def tool_augmented_agent(user_input):
  if any(op in user_input for op in ['+', '-', '*', '/']):
    return calculator_tool(user_input) # Local Tool

  # Agentic Logic
  try:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages= [
                {"role": "system", "content": "You are an assistant who uses tools if needed."},
                {"role": "user", "content": user_input}
        ]
    )
    return response.choices[0].message.content.strip()
  except Exception as e:
    return f"Error from LLM:{e}"
```

```python
# Run the tool-augmented agent
print("Tool-Augmented Agent (Calculator + GPT-4o)")
print("Ask math like '5 + 3' or natural questions like 'Who is Einstein?'. Type 'exit' to stop.")

while True:
  user_input = input("You: ").lower()

  if user_input == "exit":
    print("Agent: Shutting down.")
    break

  print("Agent: ", tool_augmented_agent(user_input))
```
