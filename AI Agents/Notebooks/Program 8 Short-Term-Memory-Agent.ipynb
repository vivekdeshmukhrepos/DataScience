```python
short_term_memory = None

def memory_agent(user_input):
  global short_term_memory

  if user_input.lower() == "what did i just say":
    if short_term_memory:
      return f"You just said:{short_term_memory}"
    else:
      return f"I havent stored anything yet"

  short_term_memory = user_input

  return "I will remember that"

```

```python
print("Short term memory agent")
print("Type Anything. Then Say: 'what did i just say'. Type 'exit' to quit")
```

```python
while True:
  user_input = input("You: ")

  if user_input.lower() == 'exit':
    print ("Agent is shutting down")
    break
  print("Agent: ", memory_agent(user_input))
```
