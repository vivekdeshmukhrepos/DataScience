```python
chat_history = []

def context_agent(user_input):
  chat_history.append(user_input)

  if "how are you" in user_input.lower():
    return "I'm doing well, Thanks!"
  elif "what we did talked about" in user_input.lower():
    if len(chat_history) >1:
      return "So far we have discussed: " + ", ".join(chat_history[:-1])
    else:
      return "We have just started chatting"
  else:
    return "Interesting, tell me more"

# Main loop
print("Context-Aware Agent")
print("Ask questions like 'How are you?', 'what we did talked about?'. Type 'exit' to stop.")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Agent: Goodbye! I’ll forget this chat.")
        break
    print("Agent:", context_agent(user_input))

```
