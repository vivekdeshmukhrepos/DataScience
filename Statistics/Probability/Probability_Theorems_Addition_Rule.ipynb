###The Addition Rule states that if one event can occur in m ways and another mutually exclusive event can occur in n ways, then the total number of possible outcomes is:
 **Total Outcomes = m + n**

**Example 1:** Color Choices

**Problem:** A student can choose 3 types of pens or 2 types of pencils. How many total choices do
they have?

**Solution:**
 • Number of pens choices = 3
 • Number of pencils choices = 2
 Total Choices = 3 + 2 = 5

```python
num_pens = 3
num_pencils = 2

# Total choices (either a pen or a pencil)
total_choices = num_pens + num_pencils

# Print the result
print(f"Total number of choices: {total_choices}")
```

**Example 2:** Course Selection

**Problem:** A student can choose from 4 science courses or 3 mathematics courses. How many total
course options are available?

**Solution:**
 • Number of science courses = 4
 • Number of mathematics courses = 3
 Total Courses = 4 + 3 = 7

```python
# Number of course options
science_courses = 4
math_courses = 3

# Total course options
total_courses = science_courses + math_courses

# Print the result
print(f"Total number of course options available: {total_courses}")
```

**Example 3:** Dice Roll or Coin Toss

**Problem:** What is the total number of outcomes if you roll a six-sided die or toss a coin?

**Solution:**
 • Possible outcomes for the die = 6
 • Possible outcomes for the coin = 2
 Total Outcomes = 6 + 2 = 8

```python
# Number of outcomes
die_outcomes = 6
coin_outcomes = 2

# Total outcomes when choosing one of the two actions (die OR coin)
total_outcomes = die_outcomes + coin_outcomes

# Print the result
print(f"Total number of possible outcomes: {total_outcomes}")
```
