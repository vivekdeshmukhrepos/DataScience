### Binomial Coefficient
The binomial coefficient represents the number of combinations of r items chosen from n items
and is often written as:

**Example 1:** Card Selection

**Problem:** How many ways can you select 5 cards from a deck of 52 cards

**Solution:**
 • Number of total cards = 52
 • Number of cards to select = 5
 (52/5) = 52!/5! x (52 - 5)!

 Step 1: Calculate factorials:
 (52/5) = 52 x 51 x 50 x 49 x 48 x … x 1/5  x 4 x 3 x 2 x 1 x 47!

 Step 2: Simplify:
 (52/5) = 311875200/120 = 2598960

```python
import math

# Values
n = 52  # total cards
r = 5   # cards to select

# Combinations: C(n, r) = n! / (r! * (n - r)!)
card_combinations = math.comb(n, r)

# Print the result
print(f"Number of ways to choose 5 cards from a deck of 52: {card_combinations}")
```

**Example 2:** Lottery Combination

**Problem:** In a lottery draw, how many ways can 6 numbers be selected from 49 numbers?

**Solution:**
 • Number of total numbers = 49
 • Number of numbers to select = 6
 (49/6) = 49!/6! x (49 - 6)!

 Step 1: Calculate factorials:
 (49/6) = 49 x 48 x 47 x 46 x 45 x 44/6 x 5 x 4 x 3 x 2 x 1

 Step 2: Simplify:
 (49/6) = 10068347520/720 = 13983816

```python
# Values
n = 49  # total numbers
r = 6   # numbers to select

# Combinations: C(n, r) = n! / (r! * (n - r)!)
lottery_combinations = math.comb(n, r)

# Print the result
print(f"Number of ways to choose 6 numbers from 49: {lottery_combinations}")
```

**Example 3:** Tournament Bracket

**Problem:** In a knockout tournament with 16 teams, how many ways can 4 teams be chosen for the
semi-final round?

**Solution:**
 • Number of total teams = 16
 • Number of teams to select = 4
 (16/4) = 16!/4! x (16 - 4)!

 Step 1: Calculate factorials:
 (16/4) = 16 x 15 x 14 x 13/4 x 3 x 2 x 1

 Step 2: Simplify:
 (16/4) = 43680/24 = 1820

```python
# Values
n = 16  # total teams
r = 4   # teams to select

# Combinations: C(n, r) = n! / (r! * (n - r)!)
semifinal_combinations = math.comb(n, r)

# Print the result
print(f"Number of ways to choose 4 teams from 16 for the semi-finals: {semifinal_combinations}")
```
