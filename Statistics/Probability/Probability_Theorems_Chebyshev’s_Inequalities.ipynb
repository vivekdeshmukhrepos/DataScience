#### Chebyshev’s Inequality
 Chebyshev’s Inequality establishes an upper bound on the probability that a random variable
deviates significantly from its mean. It applies to any distribution with a defined mean and
variance, making it a powerful tool in probability theory.

**Example 1:** Exam Scores Analysis

**Problem:** In a class of students, exam scores have a mean of 70 and a standard deviation of 10. What is
the probability that a randomly chosen student scores less than 40 or more than 100?

**Solution:**
 Step 1: Identify known values.
 • Mean μ = 70
 • Standard deviation σ = 10
 • Deviation from the mean = 30 points (either below 40 or above 100)

 Step 2: Determine k.
 k = 30/10 = 3

 Step 3: Apply Chebyshev’s Inequality.
 P(|X - 70| ≥ 30) ≤ 1/32
 P(|X - 70| ≥ 30) ≤ 1/9 ≈ 0.111

 **Conclusion:** The probability that a student scores less than 40 or more than 100 is at most 0.111 or 11.1%.

```python
# Given values
mean_score = 70  # Mean exam score
std_dev = 10     # Standard deviation of scores
deviation = 30   # Deviation from the mean (less than 40 or more than 100)
```

```python
# Calculate k (number of standard deviations away from the mean)
k = deviation / std_dev

# Apply Chebyshev's Inequality
probability_upper_bound = 1 / (k ** 2)
```

```python
# Display the result
print(f"The probability that a student scores less than 40 or more than 100 is at most: {probability_upper_bound:.3f}")
```
