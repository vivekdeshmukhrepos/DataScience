#### Combination (Order Doesn’t Matter)
 A combination is a selection of items from a larger set where order does not matter. Unlike
permutations, different sequences of the same items are considered identical in combinations.

**Example 1:** Committee Formation

**Problem:** How many ways can 3 students be chosen from a group of 5 students to form a
committee?

**Solution:**
 • Number of total students = 5
 • Number of students to select = 3
 C(5, 3) = 5!/3! x (5 - 3)!

 Step 1: Calculate factorials:
 C(5, 3) = 5 x 4 x 3 x 2 x 1/3 x 2 x 1 x 2 x 1

 Step 2: Simplify:
 C(5, 3) = 120/12 = 10

```python
import math

# Values
n = 5  # total students
r = 3  # students to select

# Combinations: C(n, r) = n! / (r! * (n - r)!)
committee_combinations = math.comb(n, r)  # Available in Python 3.8+

# Print the result
print(f"Number of ways to choose 3 students from 5: {committee_combinations}")
```

**Example 2:** Fruit Selection

**Problem:** How many ways can you select 2 fruits from a basket containing 4 different fruits?

**Solution:**
 • Number of fruits in the basket = 4
 • Number of fruits to select = 2
 C(4, 2) = 4!/2! x (4 - 2)!

 Step 1: Calculate factorials:
 C(4, 2) = 4 x 3 x 2 x 1/2 x 1 x 2 x 1

 Step 2: Simplify:
 C(4, 2) = 24/4 = 6

```python
# Values
n = 4  # total fruits
r = 2  # fruits to select

# Combinations: C(n, r) = n! / (r! * (n - r)!)
fruit_combinations = math.comb(n, r)  # Available in Python 3.8+

# Print the result
print(f"Number of ways to select 2 fruits from 4: {fruit_combinations}")
```

 **Example 3:** Team Selection

 **Problem:** In a soccer tournament, how many ways can a coach choose 4 players from a squad of
10 players?

**Solution:**
 • Number of total players = 10
 • Number of players to select = 4
 C(10, 4) = 10!/4! x (10 - 4)!

 Step 1: Calculate factorials:
 C(10, 4) = 10 x 9 x 8 x 7 x 6 x 4 x 3 x 2 x 1/4 x 3 x 2 x 1 x 6 x 5 x 4 x 3 x 2 x 1

 Step 2: Simplify:
 C(10, 4) = 5040/24 = 210

```python
# Values
n = 10  # total players
r = 4   # players to select

# Combinations: C(n, r) = n! / (r! * (n - r)!)
team_combinations = math.comb(n, r)  # Available in Python 3.8+

# Print the result
print(f"Number of ways to choose 4 players from 10: {team_combinations}")
```
