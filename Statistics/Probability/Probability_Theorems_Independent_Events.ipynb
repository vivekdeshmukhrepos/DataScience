**### Examples of Independent Events**

##  Example 1: Coin Toss and Dice Roll

#### **Problem**: What is the probability of flipping a heads on a coin toss and rolling a 3 on a six-sided die?
####**Solution**:
#### Since the coin toss and dice roll are independent P(Heads) = 1/2 and P(Rolling a 3) = 1/6
 Now calculate the combined probability:
 P(Heads and 3) = 1/2 x 1/6 = 1/12 ≈ 0.083



```python
p_heads = 1/2
p_three = 1/6

p_heads_and_three = p_heads * p_three

# Print the result
print(f"Probability of flipping heads and rolling a 3: {p_heads_and_three:.3f}")
```

#### Example 2 : Drawing Cards and Rolling
**Problem**: What is the probability of drawing a king from a standard deck of 52 cards and rolling a 6
on a six-sided die?

 **Solution**:
 • Since drawing a card and rolling a die are independent:
P(King) = 4/52 = 1/13 and P(Rolling a 6) = 1/6
 Now calculate the combined probability:
 P(King and 6) = 1/13 x 1/6 = 1/78 ≈ 0.0128

```python
p_king = 4 / 52
p_six = 1 / 6

# Since the events are independent, multiply the probabilities
p_king_and_six = p_king * p_six

# Print the result
print(f"Probability of drawing a king and rolling a 6: {p_king_and_six:.4f}")
```

 ### Example 3 : Spinning a Wheel and Drawing a Card
 **Problem:** What is the probability of spinning a wheel that lands on Red (out of 8 sections) and
drawing an Ace from a standard deck of 52 cards?

 **Solution:**
 • Since spinning the wheel and drawing a card are independent:
P(Red) = 3/8 and P(Ace) = 4/52 = 1/13
 Now calculate the combined probability:
 P(Red and Ace) = 3/8 x 1/13 = 3/104 ≈ 0.0288

```python
p_red = 3 / 8
p_ace = 4 / 52

# Since the events are independent, multiply the probabilities
p_red_and_ace = p_red * p_ace

print(f"Probability of landing on Red and drawing an Ace: {p_red_and_ace:.4f}")
```
