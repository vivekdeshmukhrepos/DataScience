#### Law of Total Probability
The Law of Total Probability is a fundamental theorem that relates the probability of an event to
the probabilities of its outcomes across different mutually exclusive and exhaustive conditions.

**Example 1:** Student Grade Prediction

**Problem:** In a university, 40% of students are from the science department, 35% are from
commerce, and 25% are from arts. The probabilities that a student passes given their department
are as follows:
 • Science students have a 70% pass rate.
 • Commerce students have a 60% pass rate.
 • Arts students have a 50% pass rate.
 What is the probability that a randomly chosen student will pass?

 **Solution:**
 Step 1: Identify the given probabilities.
 • P(Science) = 0.40
 • P(Commerce) = 0.35
 • P(Arts) = 0.25

 Step 2: Identify conditional probabilities.
 • P(Pass | Science) = 0.70
 • P(Pass | Commerce) = 0.60
 • P(Pass | Arts) = 0.50

Step 3: Apply the Law of Total Probability.
 P(Pass) = (0.70 x 0.40) + (0.60 x 0.35) + (0.50 x 0.25)

Step 4: Calculate each term.
 P(Pass) = (0.28) + (0.21) + (0.125)
 Step 5: Final Calculation.
 P(Pass) = 0.615

```python
# Given probabilities
P_science = 0.40
P_commerce = 0.35
P_arts = 0.25

P_pass_given_science = 0.70
P_pass_given_commerce = 0.60
P_pass_given_arts = 0.50
```

```python
# Apply Law of Total Probability
P_pass = (P_pass_given_science * P_science +
          P_pass_given_commerce * P_commerce +
          P_pass_given_arts * P_arts)

# Print the result
print(f"Probability that a randomly chosen student will pass: {P_pass:.3f}")
```

**Example 2**: Defective Item in a Factory

**Problem:** A factory has three production lines producing 50%, 30%, and 20% of the total output,
respectively. The defect rates for each production line are:
 • Line 1 = 3%
 • Line 2 = 5%Line 3 = 8%
 What is the probability that a randomly selected item is defective?

**Solution:**
Step 1: Identify the given probabilities.
 • P(Line 1) = 0.50
 • P(Line 2) = 0.30
 • P(Line 3) = 0.20

Step 2: Identify defect rates.
 • P(Defective | Line 1) = 0.03
 • P(Defective | Line 2) = 0.05
 • P(Defective | Line 3) = 0.08

Step 3: Apply the Law of Total Probability.
 P(Defective) = (0.03 x 0.50) + (0.05 x 0.30) + (0.08 x 0.20)

Step 4: Calculate each term.
 P(Defective) = 0.015 + 0.015 + 0.016

Step 5: Final Calculation.
 P(Defective) = 0.046


```python
# Given probabilities
P_line1 = 0.50
P_line2 = 0.30
P_line3 = 0.20

P_def_given_line1 = 0.03
P_def_given_line2 = 0.05
P_def_given_line3 = 0.08
```

```python
# Apply Law of Total Probability
P_defective = (P_def_given_line1 * P_line1 +
               P_def_given_line2 * P_line2 +
               P_def_given_line3 * P_line3)

# Print the result
print(f"Probability that a randomly selected item is defective: {P_defective:.3f}")
```

**Example 3:** Disease Diagnosis in a Population

 **Problem:** In a population:
 • 20% have Disease A
 • 30% have Disease B
 • 50% have no disease
 The probability of testing positive for each group is:
 • P(Positive | Disease A) = 0.90
 • P(Positive | Disease B) = 0.70
 • P(Positive | No Disease) = 0.10
 What is the probability that a randomly selected person tests positive?

 **Solution:**
 Step 1: Identify probabilities.
 • Probability of having Disease A: P(Disease A) = 0.20
 • Probability of having Disease B: P(Disease B) = 0.30
 • Probability of having No Disease: P(No Disease) = 0.50

 Step 2: Identify positive test rates.
 • Probability of testing positive given Disease A: P(Positive | Disease A) = 0.90
  Probability of testing positive given Disease B: P(Positive | Disease B) = 0.70
 • Probability of testing positive given No Disease: P(Positive | No Disease) = 0.10

 Step 3: Apply the Law of Total Probability.
 P(Positive) = (P(Positive | Disease A) x P(Disease A)) + (P(Positive | Disease B) x
P(Disease B)) + (P(Positive | No Disease) x P(No Disease))
 Substituting the known values:
 P(Positive) = (0.90 x 0.20) + (0.70 x 0.30) + (0.10 x 0.50)

 Step 4: Calculate each term.
 • 0.90 x 0.20 = 0.18
 • 0.70 x 0.30 = 0.21
 • 0.10 x 0.50 = 0.05
 P(Positive) = 0.18 + 0.21 + 0.05 = 0.44


```python
# Given probabilities
P_disease_A = 0.20
P_disease_B = 0.30
P_no_disease = 0.50

P_pos_given_A = 0.90
P_pos_given_B = 0.70
P_pos_given_none = 0.10
```

```python
# Apply Law of Total Probability
P_positive = (P_pos_given_A * P_disease_A +
              P_pos_given_B * P_disease_B +
              P_pos_given_none * P_no_disease)

# Print the result
print(f"Probability that a randomly selected person tests positive: {P_positive:.2f}")
```
