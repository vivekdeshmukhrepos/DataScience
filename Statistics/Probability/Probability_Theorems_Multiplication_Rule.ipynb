### Multiplication Rule with Examples
The Multiplication Rule states that if one event can occur in m ways and a second event can occur
in n ways, then the total number of outcomes for both events occurring together is:
**Total Outcomes = m × n**

**Example 1:** Outfit Combinations

**Problem**: A person has 3 shirts and 4 pants. How many different outfits can they create?

**Solution:**
 • Number of shirt choices = 3
 • Number of pant choices = 4
 Total Outfits = 3 × 4 = 12

```python
# Number of choices
num_shirts = 3
num_pants = 4

# Total outfit combinations
total_outfits = num_shirts * num_pants

# Print the result
print(f"Total number of different outfits: {total_outfits}")
```

**Example 2:** Menu Choices

**Problem:** A restaurant offers 2 appetizers, 3 main courses, and 2 desserts. How many different
meal combinations can a customer choose (one item from each category)?

**Solution:**
 • Number of appetizer choices = 2
 • Number of main course choices = 3
 • Number of dessert choices = 2

 Total Combinations = 2 × 3 × 2 = 12 **bold text**


```python
# Number of choices in each category
num_appetizers = 2
num_main_courses = 3
num_desserts = 2

# Total meal combinations
total_combinations = num_appetizers * num_main_courses * num_desserts

# Print the result
print(f"Total number of different meal combinations: {total_combinations}")
```

**Example 3:** Password Combination

**Problem:** A password consists of 2 letters (each chosen from 26 letters) followed by 3 digits (each
chosen from 0-9). How many possible passwords can be formed?

**Solution:**
 • Number of letter combinations = 26 × 26 = 676
 • Number of digit combinations = 10 × 10 × 10 = 1000

 **Total Passwords = 676 × 1000 = 676000**

```python
# Number of options
num_letters = 26      # 26 letters in the alphabet
num_digits = 10       # Digits 0 through 9

# Combinations
letter_combinations = num_letters * num_letters         # 2 letters
digit_combinations = num_digits * num_digits * num_digits  # 3 digits

# Total password combinations
total_passwords = letter_combinations * digit_combinations

# Print the result
print(f"Total number of possible passwords: {total_passwords}")
```
