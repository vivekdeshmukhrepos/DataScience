#### Permutation (Order Matters)
 A permutation refers to the arrangement of objects in a specific order. In permutations, the order
of selection is important, meaning different arrangements of the same objects are counted
separately.

**Example 1:** Seating Arrangement

 **Problem:** In how many ways can 3 students be seated in 5 chairs?

 **Solution:**
 • Number of available chairs = 5
 • Number of students to be seated = 3
 • Number of possible seating arrangements P(5, 3) = 5!/(5 - 3)!

 Calculating factorials:
 P(5,3) = 5 x 4 x 3 x 2 x 1/2 x 1 = 120/2 = 60

```python
import math
```

```python
# Values
n = 5  # total chairs
r = 3  # students to be seated

# Permutations: P(n, r) = n! / (n - r)!
seating_arrangements = math.perm(n, r)  # Available in Python 3.8+

# Print the result
print(f"Number of possible seating arrangements: {seating_arrangements}")
```

**Example 2:** Word Arrangement

**Problem:** How many different ways can the letters in the word "MATH" be arranged using exactly
3 letters?

**Solution:**
 • Number of total letters = 4
 • Number of letters to arrange = 3
 • Number of possible arrangements = P(4,3) = 4!/(4 - 3)!
 Calculating factorials:
 P(4,3) = 4 x 3 x 2 x 1/1 = 24


```python
n = 4  # total letters in the word "MATH"
r = 3  # number of letters to arrange

# Permutations: P(n, r) = n! / (n - r)!
word_arrangements = math.perm(n, r)  # Available in Python 3.8+

# Print the result
print(f"Number of different ways the letters in 'MATH' can be arranged using exactly 3 letters: {word_arrangements}")
```

 **Example 3:** Race Winners

 **Problem:** In a race with 8 participants, how many possible ways can 1st, 2nd, and 3rd place be
awarded?

**Solution:**
 • Number of participants = 8
 • Number of winners to assign = 3
 • Number of possible arrangements = P(8, 3) = 8!/(8 - 3)!
 Calculating factorials:
 P(85,3) = 8 x 7 x 6 x 5 x 4 x 3 x 2 x 1/5 x 4 x 3 x 2 x 1 = 40320/120 = 336

```python
# Values
n = 8  # total participants
r = 3  # number of winners to assign

# Permutations: P(n, r) = n! / (n - r)!
race_winners = math.perm(n, r)  # Available in Python 3.8+

# Print the result
print(f"Number of possible ways to assign 1st, 2nd, and 3rd place: {race_winners}")
```
